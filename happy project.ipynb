```python
import pandas as pd
import matplotlib as plt

#merge world happyness data
data2015=pd.read_csv("Resources/happiness/2015.csv")
data2016=pd.read_csv("Resources/happiness/2016.csv")
data2017=pd.read_csv("Resources/happiness/2017.csv")

#merge data with happyness ranking 2015/2016/2017
merge_data = pd.merge(data2015, data2016, on="Country", how="outer")
merge_data_rename = merge_data.rename(columns={"Happiness Score_x":"Happiness Score 2015",
                                               "Happiness Score_y":"Happiness Score 2016"})

final_data = pd.merge(merge_data_rename,data2017,on="Country",how="outer")
final_data_rename= final_data.rename(columns={"Happiness.Score":"Happiness Score 2017"})
#deleting data with na value and keep the columns for happiness rank 2015/2016/2017
final_data_rename=final_data_rename.dropna(how='any')
final_data_rename
```

```python
#locate the happyness score for 2015/2016/2017 and calculate the average happyness score over the 3 years
happniess_ranking=final_data_rename.loc[:,["Country","Happiness Score 2015", "Happiness Score 2016","Happiness Score 2017"]]
#add the average happyness score of each country and re-ranking
avg_happiness_score = (final_data_rename["Happiness Score 2015"]+final_data_rename["Happiness Score 2016"]+final_data_rename["Happiness Score 2017"])/3
happniess_ranking["Average Happiness Score"]=avg_happiness_score
happniess_ranking_final=happniess_ranking.sort_values("Average Happiness Score",ascending=False)

happniess_ranking_final=happniess_ranking_final.reset_index(drop=True)

#add the ranking to the table
ranking= happniess_ranking_final.index+1
happniess_ranking_final["Ranking"]=ranking
#save the final happniess_ranking_final as csv file
happniess_ranking_final.to_csv("Resources/happiness/World_Happiness_Ranking.csv",index=False)
#show the final Happiness ranking
happniess_ranking_final

```
