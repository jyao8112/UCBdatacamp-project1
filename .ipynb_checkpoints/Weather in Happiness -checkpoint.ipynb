```python
# Dependencies and Setup
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
from scipy.stats import linregress
import pandas as pd
import numpy as np
import requests
import time
import gmaps


from citipy import citipy


from config1 import api_key

from config1 import (api_key, gkey)

gmaps.configure(api_key=gkey)
```

```python
happy_matt=pd.read_csv("Resources\happiness\happiness_ranking2015.csv")
happy_matt.head()
```

```python
happy_matt_df = happy_matt[["Country","Happiness Score","Happiness Rank","Country Code"]]
happy_matt_df.head()
```

```python
lats=happy_matt["Lat"]
lngs=happy_matt["Lng"]
```

```python
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
    else:
        cities.append("no data")

# Print the city count to confirm sufficient count
len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
query_url = f"{url}appid=423160d06ea84b634174ef9038841d0f&units={units}&q="

city_data=[]
name=[]
cloudness=[]
country=[]
date=[]
humidity=[]
lat=[]
lng=[]
max_temp=[]
wind_speed=[]

print("Begining Data Retrieval")
print("---------------------------------------------------")
  
i=1
for city in cities:
        try:
            response = requests.get(query_url + city).json()
            if response['cod'] == 200:
                print("Processing Record " + str(i) + " of Set 1 | " + city)
                #city_data.append(response)
                name.append(response['name'])
                cloudness.append(response["clouds"]["all"])
                country.append(response['sys']['country'])
                date.append(response['sys']['sunrise'])
                humidity.append(response['main']['humidity'])
                lat.append(response['coord']['lat'])
                lng.append(response['coord']['lon'])
                wind_speed.append(response['wind']['speed'])
                max_temp.append(response['main']['temp_max'])
            else:
                print(f"City not found,skipping...")
                name.append("no data")
                cloudness.append("no data")
                country.append("no data")
                date.append("no data")
                humidity.append("no data")
                lat.append("no data")
                lng.append("no data")
                wind_speed.append("no data")
                max_temp.append("no data")
        except:
            print(f"{response} Exception...")
            name.append("NAN")
            cloudness.append("NAN")
            country.append("NAN")
            date.append("NAN")
            humidity.append("NAN")
            lat.append("NAN")
            lng.append("NAN")
            wind_speed.append("NAN")
            max_temp.append("NAN")

        i=i+1
```
