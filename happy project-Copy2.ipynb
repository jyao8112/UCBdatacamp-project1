```python
import pandas as pd
import matplotlib as plt
import numpy as np
import requests
import gmaps

from config import api_key
# Census & gmaps API Keys
from config1 import (api_key, gkey)


# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
#import world happyness data
happiness_data2015=pd.read_csv("Resources/happiness/2015.csv")

happiness_data2015_df=happiness_data2015.loc[:,["Country","Happiness Score","Happiness Rank"]]
happiness_data2015_df.head()
```

```python
# put city name in a list
countries=happiness_data2015_df["Country"].tolist()

lat=[]
lng=[]
shortname=[]
i=1
for country in countries:
        target_url = ('https://maps.googleapis.com/maps/api/geocode/json?''address={0}&key={1}').format(country, gkey)
        response = requests.get(target_url).json()
        print("Processing Record " + str(i) + " of Set 1 | " + country)
        shortname.append(response['results'][0]['address_components'][0]['short_name'])
        lat.append(response['results'][0]['geometry']['location']["lat"])
        lng.append(response['results'][0]['geometry']['location']["lng"])
        i=i+1
```

```python
#add lat and lng to the happyness dataframe
happiness_data2015_df["Lat"]=lat
happiness_data2015_df["Lng"]=lng
happiness_data2015_df['Short Name']=shortname
```

```python
# select top 20 happiness country and store top20 maker location for lat, lng
top20=happiness_data2015_df.loc[happiness_data2015_df["Happiness Rank"]<21]
top20_marker_locations =top20[["Lat","Lng"]]
top20_countries=top20["Country"]
top20
```

```python
# Create a map using top20 Happiness country coordinates to set markers
# Create a marker_layer using the top20 country name list to fill the info box
fig1 = gmaps.figure()
markers1 = gmaps.marker_layer(top20_marker_locations,info_box_content=[f"Country: {country}" for country in top20_countries])
fig1.add_layer(markers1)

fig1
```

```python
# select low 20 happiness country and store low 20 maker location for lat, lng
low20=happiness_data2015_df.loc[happiness_data2015_df["Happiness Rank"]>137]
low20_marker_locations =low20[["Lat","Lng"]]
low20_countries=low20["Country"]
```

```python
# Create a map using low20 Happiness country coordinates to set markers
# Create a marker_layer using the low20 country name list to fill the info box
fig2 = gmaps.figure()
markers2 = gmaps.marker_layer(low20_marker_locations,info_box_content=[f"Country: {country}" for country in low20_countries])
fig2.add_layer(markers2)
fig2
```

```python
#import world current health expenditure data set from csv file and organized the data for analysis
health_current_expenditure = pd.read_csv("Resources/current_expenditure_Data.csv")
health_current_expenditure_2015=health_current_expenditure.loc[:,["Country Name","2015 [YR2015]"]]
health_current_expenditure_2015=health_current_expenditure_2015.rename(columns={"2015 [YR2015]":"2015 Current Health Expediture (%GDP)",
                                                                       "Country Name":"Country"})
#import world domestic govement health expenditure data set from csv file and organized the data for analysis
health_gov_expenditure=pd.read_csv("Resources/Domestic_Govement_Health_Expenditure Data.csv")
health_gov_expenditure_2015=health_gov_expenditure.loc[:,["Country Name","2015 [YR2015]"]]
health_gov_expenditure_2015=health_gov_expenditure_2015.rename(columns={"2015 [YR2015]":"2015 govement Health Expediture (%GDP)",
                                                                       "Country Name":"Country"})
```

```python
merge1=happiness_data2015_df.merge(health_current_expenditure_2015,on='Country',how="left")
merge2=merge1.merge(health_gov_expenditure_2015,on='Country',how="left")

merge2.count()
```

```python
#import world out of pocket expediture
import json 

#oop_url="http://apps.who.int/gho/athena/data/GHO/GHED_OOP_pc_US_SHA2011.json?profile=simple&filter=COUNTRY:*;YEAR:2015"
oop_url="http://apps.who.int/gho/athena/data/GHO/GHED_OOPSCHE_SHA2011.json?profile=simple&filter=COUNTRY:*;YEAR:2015"
response_OOP = requests.get(oop_url).json()

total=np.arange(0,194,1)
oop_value=[]
oop_country=[]

i=0
for i in total:
    oop_country.append(response_OOP['fact'][i]['dim']["COUNTRY"])
    oop_value.append(response_OOP['fact'][i]['Value'])
    i=i+1
```

```python
#Merge oop_data to final dataframe
oop_data=pd.DataFrame({"Country":oop_country,
                       "OOP/CHE(%)":oop_value})
final_health_data=merge2.merge(oop_data,on='Country',how="left")
final_health_data
```
